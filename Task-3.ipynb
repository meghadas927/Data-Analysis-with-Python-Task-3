```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data =pd.read_csv("C:\\Internships\\Main flow and service\\Data analysis with python internship\\Task-3\\householdtask3.csv")
```

```python
Data.head(10)
```

```python
# Scatter plot with year against own
plt.scatter(Data["year"],Data["own"])
# Adding title to the plot
plt.title("Scatter plot")
#Setting the x and Y labels
plt.xlabel("year")
plt.ylabel("Own")
#Showing the Chart
plt.show()
```

```python
# Line chart with year against own
plt.plot(Data["year"])
plt.plot(Data["own"])
# Adding title to the plot
plt.title("Line chart")
#Setting the x and Y labels
plt.xlabel("year")
plt.ylabel("Own")
#Showing the Chart
plt.show()
```

```python
# Bar chart or bar plot with year against own
plt.bar(Data["year"],Data["own"])
# Adding title to the plot
plt.title("Bar chart")
#Setting the x and Y labels
plt.xlabel("year")
plt.ylabel("Own")
#Showing the Chart
plt.show()
```

```python
# Histogram
plt.hist(Data["income"])
plt.title("Histogram")
plt.show()
```
